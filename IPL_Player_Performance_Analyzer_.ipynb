Analyzed batting stats to identify top performers by Runs, Strike Rate, and Batting Average.

```python
# Importing essential libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style
sns.set(style="whitegrid")

from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('most_runs_average_strikerate (5).csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
top_run_scorers = df.sort_values(by='total_runs',ascending=False).head(10)
top_run_scorers[['batsman','total_runs']]
```

```python
top_strike_rate = df[df['numberofballs'] > 100].sort_values(by='strikerate', ascending=False).head(10)
top_strike_rate[['batsman', 'strikerate']]
```

```python
top_average = df[df['out'] > 10].sort_values(by='average', ascending=False).head(10)
top_average[['batsman', 'average']]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='total_runs', y='batsman', data=top_run_scorers, palette='viridis')
plt.title('Top 10 IPL Run Scorers', fontsize=16)
plt.xlabel('Total Runs')
plt.ylabel('Player')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='strikerate', y='batsman', data=top_strike_rate, palette='rocket')
plt.title('Top 10 IPL Strike Rates (Min 100 Balls)', fontsize=16)
plt.xlabel('Strike Rate')
plt.ylabel('Player')
plt.show()
```
